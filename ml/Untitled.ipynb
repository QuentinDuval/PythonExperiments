```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

xs = np.arange(-5, 5, 0.1)
ys = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(xs, ys)
Z = 3*X**2 + 6*Y**2 + 8*X*Y

plt.contour(X, Y, Z, 20)
plt.show()
```

```python
def euler(x: float, y: float, h: float):
    xs = [x]
    ys = [y]
    while xs[-1] < 4:
        d = xs[-1] ** 2 - ys[-1] ** 2
        ys.append(ys[-1] + d * h)
        xs.append(xs[-1] + h)
    return xs, ys

plt.figure(figsize=(12,8))
for h in np.arange(0.1, 0.9, 0.02):
    plt.plot(*euler(0, 1, h))
plt.show()
```

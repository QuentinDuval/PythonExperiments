```python
"""
Displaying the K-Means:
- showing the centroids
- showing the lines between the centroids
"""

data = test_huge(size=10_000)
points = data['inputs']
centroids = np.stack(data['outputs'])

fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:,0], points[:,1], points[:,2], alpha=0.01)
ax.scatter(centroids[:,0], centroids[:,1], centroids[:,2], color='red')
ax.plot(centroids[:2,0], centroids[:2,1], centroids[:2,2], color='orange')
ax.plot(centroids[1:,0], centroids[1:,1], centroids[1:,2], color='orange')
ax.plot(centroids[(0,2),0], centroids[(0,2),1], centroids[(0,2),2], color='orange')
plt.show()
```

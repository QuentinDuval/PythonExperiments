```python
"""
Approximation of differential equation:

dx / dt = 0.1 (flow in) - x / 80 (flow out)
"""

%matplotlib inline
import matplotlib
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np

def solution(ts, start_point, start_time=0):
    return 8 - (8 - start_point) * np.exp(- (np.array(ts) - start_time) / 80)

def simulation(start_point):
    def derivative(x):
        return 0.1 - x / 80
    
    xs = [start_point]
    ts = [0]
    step = 0.01
    for _ in range(int(500 / step)):
        ts.append(ts[-1] + step)
        xs.append(xs[-1] + step * derivative(xs[-1]))
        
    fig, ax = plt.subplots(figsize=(15, 10))
    ax.plot(ts, xs)
    ax.plot(ts, solution(ts, start_point))
    plt.show()
    
simulation(start_point=1)
```

```python
end_time = 500

fig, ax = plt.subplots(figsize=(15, 10))
for start_time in np.arange(0, 200, 20):
    for start_point in np.arange(1, 15, 1):
        ts = np.arange(start_time, end_time)
        ax.plot(ts, solution(ts, start_point, start_time))
plt.show()
```

```python
%matplotlib inline
import matplotlib
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0, 8, 0.01)
x = np.cos(2*t) + 3*np.sin(6*t)
y = np.sin(3*t) + 2*np.sin(5*t)

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x, y, lw=3)
plt.show()
```

```python
import math

pi = math.pi

def butterfly(t):
    amplitude = np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12) ** 5
    x = np.sin(t) * amplitude
    y = np.cos(t) * amplitude
    return x, y

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(*butterfly(np.arange(0, 4*pi, 0.01)), lw=2)
ax.plot(*butterfly(np.arange(4*pi, 8*pi, 0.01)), lw=2)
ax.plot(*butterfly(np.arange(8*pi, 10*pi, 0.01)), lw=2)
plt.show()
```
